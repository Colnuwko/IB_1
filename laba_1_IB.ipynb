Закодировать любой текст (не менее 500 символов), любым произвольным шифром простой подстановки (моноалфавитная замена) или перестановки

```python
import os
from tqdm import trange
print("heum")
Alfavit_low = ('абвгдеёжзийклмнопрстуфхцчшщъыьэюя')
```

Создадим ключ шифрования

```python
length = len(Alfavit_low) - 1

shift = 4

Alfavit_key = Alfavit_low[+shift:] + Alfavit_low[:+shift]

print(Alfavit_low)
print(Alfavit_key)
```

```python
with open("text.txt", 'r', encoding="utf-8") as f:
    
    text = f.read()
    print(text)
```

Преобразуем на текст в нормальный вид для дальнейшей работы.

```python
text = text.lower()
print(len(text)-1)
#last element 4750

```

Теперь последовательно будем менять буквы из первого алфавита на буквы из второго


```python
result_text = ''
for i in trange(len(text)-1):
    index_symbol = str(Alfavit_low).find(text[i])
    if  index_symbol != -1:
        result_text = result_text + Alfavit_key[index_symbol]
    else:
        result_text = result_text + text[i]
print(result_text)
```

Осталось записать это в файл


```python
with open('text_shifr.txt', 'w+', encoding="utf-8") as file:
    file.write(result_text)
```

Запишем в файл ключ

```python
with open('key_of_shifr.txt', 'w+', encoding="utf-8") as file:
    file.write(str(Alfavit_low)+"->"+str(Alfavit_key))
```
