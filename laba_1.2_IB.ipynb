Сделаем частотный анализ по нашему тексту

```python
dic = dict()
text = open("Beleberda.txt", "r", encoding="utf-8").read()
lst = list(set(text))
l = len(text)
for i in lst:
    dic[i] = (text.count(i))/l
dic = sorted(dic.items(), key=lambda x: x[1], reverse=True)
print(dic)
```
